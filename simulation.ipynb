# モデルの式一覧
- $p=\frac{(1+\nu_1+\nu_2\frac{L_{f-1}}{C_{-1}+G_{-1}})(W_{-1}+T_{v-1}+T_{c-1})}{u^T\gamma_1 k_{-1}}$
  - 価格を安定させたがること（塩沢）を反映した価格設定式はないか？
  - マークアップ率は価格競争と投資資金回収率を主な引数とする関数で書かれるのでは？集計的SFCモデルでは投資資金回収率に比例する項と定数項の和が価格マークアップの水準になるのが現実的か？
- $T_i=\tau_1 W_{-1}$
- $T_a=\tau_2(E_{h-1}+M_{h-1}-L_{h-1})$
- $T_v=\tau_3(C_{-1}+I_{-1}+G_{-1})$
- $T_c=\tau_4(C_{-1}+G_{-1}+I_{-1}-W_{-1}-T_v)$
- $G=(1+\beta)G_{-1}$
- $g=\frac{G}{p}$
- $C=\alpha_1(W_{-1}-T_a-T_i-r_L L_{-1}+P_{h-1}+S_{-1}) + \alpha_2(E_{h-1}+M_{-1}-L_{-1})$
- $c=\frac{C}{p}$
- $A=A_{-1}(1 + \mu_1 + \mu_2 \frac{i_{-1}}{k_{-1}})$
  - 技術水準上昇率が投資の一次関数として決まる。
- $u = \frac{c+i+g}{\gamma_1 k_{-1}}$
- $i=\delta k_{-1} + (u_{-1}-u^T)\gamma_2 k_{-1} + \gamma_3\frac{M_{f-1}-L_{f-1}}{p}$
- $I=p i$
- $k=(1-\delta)k_{-1}+i$
- $K=p k$
- $v=u \frac{k_{-1}}{A}$
  - 労働稼働率は、技術水準（と労働者数の積）/消費および投資の実質需要　で定義する。
- $w=(1-\zeta_1)w_{-1} + \zeta_1 w_{-1}\{1 + \zeta_2(v - v^T)\}$
- $W=w(c_{-1}+g_{-1}+i_{-1})$
  - 労働者側の賃金交渉力が高いときは名目賃金が上昇しやすく、逆に労働者側の賃金交渉力が低いときは名目賃金が上がりにくい。労働者側の賃金交渉力の相対的な強さは、労働稼働率の一次関数で決まることにする。
- $P=C+G+I-W-T_c-T_v-r_L L_{f-1}$
- $P_h=\max\{0, \theta_1(P-I)+\theta_2(M_{f-1}-L_{f-1})\}\frac{e_{h-1}}{e_{-1}}$
- $P_b=\max\{0, \theta_1(P-I)+\theta_2(M_{f-1}-L_{f-1})\}\frac{e_{b-1}}{e_{-1}}$
- $P_f=P-P_h-P_b$
- $S=\theta_3(r_L L_{-1}+P_b)+\theta_4 E_{b-1}$
- $NL_h=-C+W-T_i-T_a-r_L L_{h-1}+P_h+S$
- $NL_f=-I+P_f$
- $NL_b=r_L L_{-1}+P_b-S$
- $NL_g=-G+T_i+T_a+T_v+T_c$
- $L_h=\epsilon_1 NL_h+\epsilon_2 C$
- $\Delta L_h=L_h-L_{h-1}$
- $\Delta L_f=\max\{-L_{f-1}, (\lambda_3 + \lambda_4(\frac{P - P_f}{E_{h-1} + E_{b-1}} - r_L))(I+W+T_v+T_c+r_L L_{f-1} - \phi M_{f-1})\}$
  - もっとリアルな貸付金水準の行動方程式はないか？要調査
- $L_f=L_{f-1}+\Delta Lf$
- $L=L_h+L_f$
- $\Delta L=L-L_{-1}$
- $\Delta e=\frac{1}{p_{-1}}(1-\lambda_3 - \lambda_4(\frac{P - P_f}{E_{h-1} + E_{b-1}} - r_L))(I+W+T_v+T_c+r_L L_{f-1} - \phi M_{f-1})$
- $E=E_{-1}+p_{e-1}\Delta e$
- $e=e_{-1}+\Delta e$
- $E_h=(\lambda_1+\lambda_2\frac{P_{h-1}}{E_{h-1}})(NL_h+\Delta L_h+E_{h-1}+M_{h-1})$
- $E_b=(1-\lambda_5 r_L+\lambda_5 \frac{P_{b-1}}{E_{b-1}})(NL_b+E_{b-1})$
- $p_e=\frac{E_h+E_b}{e}$
- $e_h=\frac{E_h}{p_e}$
- $e_b=\frac{E_b}{p_e}$
- $\Delta e_h=e_h-e_{h-1}$
- $\Delta e_b=e_h-e_{b-1}$
- $\Delta M_h=NL_h-p_{e-1}\Delta e_h+\Delta L_h$
- $M_h=M_{h-1}+\Delta M_h$
- $\Delta M_f=NL_f+\Delta L_f+p_{e-1}\Delta e$
- $M_f=M_{f-1}+\Delta M_f$
- $M=M_h+M_f$
- $\Delta M=M-M_{-1}$
- $\Delta H = NL_b - p_{e-1}\Delta e_b + \Delta M - \Delta L$
- $H = H_{-1} + \Delta H$
- $NW_h=M_h-L_h+E_h$
- $NW_f=K+M_f-L_f-E$
- $NW_b=-M+L+E_b+H$
- $NW_g=H$
- $D=-E+E_h+E_b$
- $NW=NW_h+NW_f+NW_b+NW_g$

隠された恒等式として

- $K+D-NW=0$
- $NL_h+NL_f+NL_b+NL_g=0$
- $\Delta L=\Delta L_h+\Delta L_f$
- $\Delta M=\Delta M_h+\Delta M_f$
- $NL_g+\Delta H=0$

を使ってモデルの会計的一貫性を確認する

# モデルを作って実行

```julia
# シミュレーション期間
N = 200
# 変数定義
p, pe, A, u, v, W, w, rL = zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), 0.05*ones(N)
Ti, Ta, Tv, Tc = zeros(N), zeros(N), zeros(N), zeros(N)
C, G, I, c, g, i = zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N)
K, k = zeros(N), zeros(N)
P, Pf, Ph, Pb, S = zeros(N), zeros(N), zeros(N), zeros(N), zeros(N)
NLh, NLf, NLb, NLg = zeros(N), zeros(N), zeros(N), zeros(N)
Lh, Lf, L, ΔLh, ΔLf, ΔL = zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N)
Mh, Mf, M, ΔMh, ΔMf, ΔM = zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N)
Eh, Eb, E, eh, eb, e, Δeh, Δeb, Δe = zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N)
H, ΔH = zeros(N), zeros(N)
D, NW, NWh, NWf, NWb, NWg = zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N)

uT, vT = 0.8, 0.8
α1, α2 = 0.8, 0.02
β = 0.02
γ1, γ2, γ3 = 1.0, 0.5, 0.2
δ = 0.05
ϵ1, ϵ2 = 1.0, 1.0
τ1, τ2, τ3, τ4 = 0.3, 0.02, 0.1, 0.2
θ1, θ2, θ3, θ4 = 0.2, 0.1, 0.2, 0.1
μ1, μ2 = 0.0, 0.1
ν1, ν2 = 0.2, 0.2
ζ1, ζ2 = 0.5, 0.5
λ1, λ2, λ3, λ4, λ5 = 0.5, 1.0, 0.5, 5.0, 2.0
ϕ = 0.5

# 初期値
k[1], Mh[1], Mf[1], Lh[1], Lf[1] = 2.0, 1.0, 1.0, 1.0, 1.0
Eh[1], E[1], Eb[1] = 1.0, 2.0, 1.0
p[1], pe[1], A[1] = 1.0, 1.0, 2.0
H[1] = 1.0
K[1], eh[1], eb[1], e[1] = p[1]*k[1], Eh[1]/pe[1], Eb[1]/pe[1], E[1]/pe[1]
M[1], L[1], D[1] = Mh[1] + Mf[1], Lh[1] + Lf[1], Eh[1] - E[1] + Eb[1]
NWh[1] = Mh[1] - Lh[1] + Eh[1]
NWf[1] = K[1] + Mf[1] - Lf[1] - E[1]
NWb[1] = -M[1] + L[1] + Eb[1] + H[1]
NWg[1] = H[1]
NW[1] = K[1] + D[1]

W[1], C[1], G[1], I[1] = 1.0, 1.0, 1.0, 0.0
c[1], g[1], i[1], w[1] = C[1]/p[1], G[1]/p[1], I[1]/p[1], W[1]
Ti[1] = τ1*W[1]
Ta[1] = τ2*(Eh[1] + Mh[1] - Lh[1])
Tv[1] = τ3*(C[1] + I[1] + G[1])
Tc[1] = τ4*(C[1] + G[1] + I[1] - W[1] - Tv[1])
u[1], v[1] = (c[1] + g[1])/(γ1*k[1]), A[1]/(c[1]+g[1]+i[1])
P[1] = C[1] + G[1] + I[1] - W[1] - Tc[1] - Tv[1] - rL[1]*Lf[1]
Ph[1] = max(0, θ1*(P[1] - I[1]) + θ2*(Mf[1] - Lf[1]))*eh[1]/e[1]
Pb[1] = max(0, θ1*(P[1] - I[1]) + θ2*(Mf[1] - Lf[1]))*eb[1]/e[1]
Pf[1] = P[1] - Ph[1] - Pb[1]
S[1] = θ3*(rL[1]*L[1] + Pb[1]) + θ4*(L[1]+Eb[1])
NLh[1] = -C[1] + W[1] - Ti[1] - Ta[1] - rL[1]*Lh[1] + Ph[1] + S[1]
NLf[1] = -I[1] + Pf[1]
NLg[1] = -G[1] + Ti[1] + Ta[1] + Tv[1] + Tc[1]


# シミュレーション
function run(RANGE)
    for t = RANGE
        p[t] = (1 + ν1 + ν2*Lf[t-1]/(C[t-1] + G[t-1]))*(W[t-1] + Tv[t-1] + Tc[t-1])/(uT*γ1*k[t-1])
        Ti[t] = τ1*W[t-1]
        Ta[t] = τ2*(Eh[t-1] + Mh[t-1] - Lh[t-1])
        Tv[t] = τ3*(C[t-1] + I[t-1] + G[t-1])
        Tc[t] = τ4*(C[t-1] + G[t-1] + I[t-1] - W[t-1] - Tv[t])
        G[t] = (1 + β)*G[t-1]
        g[t] = G[t]/p[t]
        C[t] = α1*(W[t-1] - Ta[t] - Ti[t] - rL[t]*L[t-1] + Ph[t-1] + S[t-1]) + α2*(Eh[t-1] + M[t-1] - L[t-1])
        c[t] = C[t]/p[t]
        A[t] = A[t-1]*(1 + μ1 + μ2*i[t-1]/k[t-1])
        i[t] = max(0, δ*k[t-1] + min((u[t-1] - uT)*γ2*k[t-1], γ3*(Mf[t-1] - Lf[t-1])/p[t]))
        u[t] = (c[t] + i[t] + g[t])/(γ1*k[t-1])
        I[t] = p[t]*i[t]
        k[t] = (1 - δ)*k[t-1] + i[t]
        K[t] = p[t]*k[t]
        v[t] = u[t]*(k[t-1]/A[t]) # 資本の量/フル労働によって動かせる資本の量=資本稼働率１００%のときの労働稼働率
        w[t] = (1 - ζ1)*w[t-1] + ζ1*w[t-1]*(1 + ζ2*(v[t] - vT))
        W[t] = w[t]*(c[t-1] + g[t-1] + i[t-1])
        P[t] = C[t] + G[t] + I[t] - W[t] - Tc[t] - Tv[t] - rL[t]*Lf[t-1]
        Ph[t] = max(0, θ1*(P[t] - I[t]) + θ2*(Mf[t-1] - Lf[t-1]))*eh[t-1]/e[t-1]
        Pb[t] = max(0, θ1*(P[t] - I[t]) + θ2*(Mf[t-1] - Lf[t-1]))*eb[t-1]/e[t-1]
        Pf[t] = P[t] - Ph[t] - Pb[t]
        S[t] = θ3*(rL[t]*L[t-1] + Pb[t]) + θ4*Eb[t-1]
        NLh[t] = -C[t] + W[t] - Ti[t] - Ta[t] - rL[t]*Lh[t-1] + Ph[t] + S[t]
        NLf[t] = -I[t] + Pf[t]
        NLb[t] = rL[t]*L[t-1] + Pb[t] - S[t]
        NLg[t] = -G[t] + Ti[t] + Ta[t] + Tv[t] + Tc[t]
        Lh[t] = ϵ1*NLh[t] + ϵ2*C[t]
        ΔLh[t] = Lh[t] - Lh[t-1]
        ΔLf[t] = max(-Lf[t-1], (λ3 + λ4*((P[t] - Pf[t])/(Eh[t-1] + Eb[t-1]) - rL[t]))*(I[t]+W[t]+Tv[t]+Tc[t]+rL[t]*Lf[t-1] - ϕ*Mf[t-1]))
        Lf[t] = Lf[t-1] + ΔLf[t]
        L[t] = Lh[t] + Lf[t]
        ΔL[t] = L[t] - L[t-1]
        Δe[t] = (1-λ3 - λ4*((P[t] - Pf[t])/(Eh[t-1] + Eb[t-1]) - rL[t]))*(I[t]+W[t]+Tv[t]+Tc[t]+rL[t]*Lf[t-1] - ϕ*Mf[t-1])/pe[t-1]
        E[t] = E[t-1] + pe[t-1]*Δe[t]
        e[t] = e[t-1] + Δe[t]
        Eh[t] = (λ1 + λ2*Ph[t-1]/Eh[t-1])*(NLh[t] + ΔLh[t] + Eh[t-1] + Mh[t-1])
        Eb[t] = (1 - λ5*rL[t] + λ5*Pb[t-1]/Eb[t-1])*(NLb[t] + Eb[t-1])
        pe[t] = (Eh[t] + Eb[t])/e[t]
        eh[t] = Eh[t]/pe[t]
        eb[t] = Eb[t]/pe[t]
        Δeh[t] = eh[t] - eh[t-1]
        Δeb[t] = eb[t] - eb[t-1]
        ΔMh[t] = NLh[t] - pe[t-1]*Δeh[t] + ΔLh[t]
        Mh[t] = Mh[t-1] + ΔMh[t]
        ΔMf[t] = NLf[t] + ΔLf[t] + pe[t-1]*Δe[t]
        Mf[t] = Mf[t-1] + ΔMf[t]
        M[t] = Mh[t] + Mf[t]
        ΔM[t] = M[t] - M[t-1]
        ΔH[t] = NLb[t] - pe[t-1]*Δeb[t] + ΔM[t] - ΔL[t]
        H[t] = H[t-1] + ΔH[t]
        NWh[t] = Mh[t] - Lh[t] + Eh[t]
        NWf[t] = K[t] + Mf[t] - Lf[t] - E[t]
        NWb[t] = -M[t] + L[t] + Eb[t] + H[t]
        NWg[t] = -H[t]
        D[t] = -E[t] + Eh[t] + Eb[t]
        NW[t] = NWh[t] + NWf[t] + NWb[t] + NWg[t]
    end
end
run(2:N)
```

# 会計的一貫性の確認

```julia
plot(2:N, NLh[2:end] + NLf[2:end] + NLb[2:end] + NLg[2:end], label="NLh + NLf + NLb + NLg")
plot!(2:N, K[2:end] + D[2:end] - NW[2:end], label="K + D - NW")
plot!(2:N, ΔL[2:end] - ΔLh[2:end] - ΔLf[2:end], label="ΔL - ΔLh - ΔLf")
plot!(2:N, ΔM[2:end] - ΔMh[2:end] - ΔMf[2:end], label="ΔM - ΔMh - ΔMf")
plot!(2:N, NLg[2:end] + ΔH[2:end], label="NLg + ΔH")
```
